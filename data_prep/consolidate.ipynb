```python
import pandas as pd
```

```python
well_code_list = [7405, 7078, 5599, 5351, 7289]

df_list = [pd.read_csv(f'../datasets/well_data/well_{well_code}_data.csv') for well_code in well_code_list]

df = pd.concat(df_list)
```

```python
df
```

```python
fluid_list = ['OIL', 'GAS']

for fluid in fluid_list:
    fluid_df = df[['DATEPRD', f'BORE_{fluid}_VOL']].groupby('DATEPRD').sum()
    fluid_df.to_csv(f'../datasets/consolidated_data/{fluid.lower()}_data.csv')
```
